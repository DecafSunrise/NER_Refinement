```python
import spacy
from spacy import displacy

import requests
import boilerpy3
from boilerpy3 import extractors
```

```python
nlp = spacy.load("en_core_web_sm")
nlpLg = spacy.load("en_core_web_lg")
nlpTrf = spacy.load("en_core_web_trf")
```

```python
extractor = extractors.ArticleExtractor()
```

```python
url = r"https://www.nbcnews.com/news/us-news/biden-diversify-usps-board-cement-democrats-oversight-appointments-n1258767"
```

```python
doc = extractor.get_doc_from_url(url)
content = doc.content
title = doc.title
```

```python
title
```

```python
print(content)
```

```python
doc2 = nlp(content)
print(f"Small model - {len(doc2.ents)} ents")
```

```python
doc3 = nlpLg(content)
print(f"Large model - {len(doc3.ents)} ents")
```

```python
doc4 = nlpTrf(content)
print(f"Transformers model - {len(doc4.ents)} ents")
```

```python
displacy.render(doc4, style="ent")
```
